```python
%matplotlib inline
import torch
import matplotlib.pyplot as plt
from IPython import display
```

```python
def set_svg_figure():
  display.set_matplotlib_formats('svg')
def set_figsize(figsize=(3.5,2.5)):
  set_svg_figure()
  plt.rcParams['figure.figsize']=figsize
```

```python
def xy_plot(x_vals,y_vals,name):
  set_figsize(figsize=(5,2.5))
  plt.plot(x_vals.detach().numpy(),y_vals.detach().numpy())
  plt.xlabel('X')
  plt.ylabel(name+'(x)')
```

```python
x = torch.arange(-8.0,8.0,0.01,requires_grad=True)
y = x.relu()
xy_plot(x,y,'relu')
```

```python
y.sum().backward()
xy_plot(x,x.grad,'Grad of relu')
```

```python
y = x.sigmoid()
xy_plot(x,y,'sigmoid')
```

```python
x.grad.zero_()
y.sum().backward()
xy_plot(x,x.grad,'sigmoid of grad')
```
